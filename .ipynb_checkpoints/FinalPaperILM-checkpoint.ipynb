# DS-SEA-3 final paper: Do demographics predict healthcare spending?
    

### Problem statement and hypothesis

The USA government’s expenditure on healthcare is set to reach 20 % of GDP by 2025<sup>1</sup>. There are many contributing factors, but is there some way to target efforts to reduce spending?

Examining the specific geographical areas in the Centers for Medicare and Medicaid (CMS) spending data should reveal differences in spending between areas. Those differences should be able to be correlated with demographic features of those areas from US decennial census data via regression modelling. 

It is expected that the percentage of elderly people living in an area (75+ rears old) will be positively correlated with increased spending in an area. There may be other correlations as well.


### Description of your data set and how it was obtained

The dataset is made up of three different data sources:

1. Centers for Medicaid and Medicare outpatient spending data (2011-2014)<sup>2</sup>
2. USA Census bureau 2010 decennial census data<sup>3</sup>
3. US Dept of Housing and and Urban Development HUD USPS ZIP Code crosswalk Q2 2016<sup>4</sup>.

The CMS spending data was downloaded directly as a flat file CSV. As was the the ZIP to county crosswalk table.

The 2010 Decennial census data was only available via the census API and suitable features were accessed using Laura Kurup’s Census API Python Script<sup>5</sup>. The script was configured with a census user API key, and the location_type variable set to ‘county’. The variables requested are listed out in the census_variables.csv file. There are 78 in total covering male and female age ranges, total population, racial and ethnic demography, plus household size.

### Data Cleaning

Create one data frame out of the three starting dataframes. To achieve that several steps are performed:

1. Ensure that all the zip codes still have leading zeros
2. Payment info is grouped and summed into one payment value per zip code (rather than several payment types)
3. create combined FIPS codes from component state and county FIPS to match counties between dataframes
4. change column names that will be used as keys with the ```merge``` command. The dataframes to be merged must each have a column with the same name which will be used to line up each row from each dataframe.
5. merge the zip to county cross walk and census data frames in a new dataframe
6. merge zip-census data frame with the payment dataframe
7. Drop columns that are 100 % NaN values
8. convert columns we want to treat as numeric to numeric types
8. drop rows with NaN values
7. calculate adjusted payment for zip codes not entirely within counties

````python
#ENSURE LEADING ZEROS ARE PRESENT ON ALL ZIP CODES IN PAYMENT DATA
rawPaymentData['Provider_Zip_Code'] = rawPaymentData['Provider_Zip_Code'].apply(lambda x: x.zfill(5))

#SIMPLIFY PAYMENT INFO DOWN TO SUMMED PAYMENTS PER ZIP CODE
rawPaymentData['Total'] = rawPaymentData['Outpatient_Services'] * rawPaymentData['Average_Total_Payments'] 
paymentData = rawPaymentData[['Provider_Zip_Code','Total']].groupby(by='Provider_Zip_Code', as_index=False).sum()

#CREATE FIPS CODE IN CENSUSDATA DATAFRAME - ADD BACK LEADING ZEROS AND CONCATENATE
censusData['state_fips'] = censusData['state_fips'].apply(lambda x: x.zfill(2))
censusData['county_fips'] = censusData['county_fips'].apply(lambda x: x.zfill(3))
censusData['FIPS'] = censusData['state_fips']+censusData['county_fips']

#CHANGE COLUMN NAMES TO MERGE BY THOSE COLUMN NAMES
zipCountyCrosswalk.rename(columns={'COUNTY':'FIPS'}, inplace=True)
paymentData.rename(columns={'Provider_Zip_Code':'ZIP'}, inplace=True)

#MERGE CENSUS AND ZIP CROSSWALK DATA FRAME
censusZip = pd.merge(censusData, zipCountyCrosswalk, on = 'FIPS', how = 'left')

#MERGE THE CENSUSZIP DATAFRAME AND THE CMS PAYMENT DATA DATAFRAME
censusZipPayment = pd.merge(censusZip, paymentData, on = 'ZIP', how = 'left')
````

### Feature engineering - trying to improve the model

The state a zip code is in may play a role in the expenditure in that zip code. Dummy features can be made for each state and those features can then be used for fitting models.

````python
paymentStates = pd.get_dummies(numericCensusZipPaymentNoNanAgain.state, prefix='state')
paymentStates.drop(paymentStates.columns[0], axis=1, inplace=True)

# concatenate the original DataFrame and the dummy DataFrame
numericCensusZipPaymentNoNanAgainStates = pd.concat([numericCensusZipPaymentNoNanAgain, paymentStates], axis=1)
````

````python
#set the features to be just the states - i.e. where a zip code is in at state
featurelist = paymentStates.columns.values

#set up x and y
y = numericCensusZipPaymentNoNanAgainStates['adjTotal']
X = numericCensusZipPaymentNoNanAgainStates[featurelist]

#set up a linear regresiion estimateor and run cross validation to return rmse
linreg = LinearRegression()
msescores = cross_val_score(linreg,X,y,cv=10,scoring='mean_squared_error')
linregrmsescores = np.mean(np.sqrt(-msescores))

#set up a decision tree estimateor and run cross validation to return rmse
treereg = DecisionTreeRegressor()
msescores = cross_val_score(treereg,X,y,cv=10,scoring='mean_squared_error')
treeregrmsescores = np.mean(np.sqrt(-msescores))

#print results
print 'nullPredictionRMSE: ', nullPredictionRMSE #the guessing value. 
print 'linreg rmsescores: ', linregrmsescores #wayway worse than guessing
print 'treereg rmsescores:', treeregrmsescores #only sligthly better than guessing 
````

This approach worked better for the decision tree model, now slightly better than guessing, but orders of magnitude worse for the linear regression. You might expect this when using only categorical information that doesn't have any direct linear relationship.

### Trying a different model - Random Forest

````python
#DEFINE FEATURE SET USING RACE AND AGE FEATURES
feature_cols=['White alone_2010',
'Black or African American alone_2010',
'American Indian and Alaska Native alone_2010', 'Asian alone_2010',
'Native Hawaiian and Other Pacific Islander alone_2010',
'Some Other Race alone_2010', 'Two or More Races_2010',
'Not Hispanic or Latino_2010', 'Hispanic or Latino_2010',
'Male: !! 65 and 66 years_2010', 'Male: !! 67 to 69 years_2010',
'Male: !! 70 to 74 years_2010', 'Male: !! 75 to 79 years_2010',
'Male: !! 80 to 84 years_2010', 'Male: !! 85 years and over_2010',
'Female: !! 85 years and over_2010', 'Female: !! Under 5 years_2010',
'Female: !! 65 and 66 years_2010', 'Female: !! 67 to 69 years_2010',
'Female: !! 70 to 74 years_2010', 'Female: !! 75 to 79 years_2010',
'Female: !! 80 to 84 years_2010']

#SETUP X AND y
y = numericCensusZipPaymentNoNanAgainStates['adjTotal']
X = numericCensusZipPaymentNoNanAgainStates[feature_cols]

#IMPORT RANDOM FOREST REGRESSOR
from sklearn.ensemble import RandomForestRegressor

#INSTANTIATE THE ESTIMATOR
rfreg = RandomForestRegressor()

#DETERMINE RMSE
from sklearn.cross_validation import cross_val_score
MSE_scores = cross_val_score(rfreg, X, y, cv=10, scoring='mean_squared_error')
RMSE_scores = np.mean(np.sqrt(-MSE_scores))
print 'random forest RMSE_scores: ', RMSE_scores

#DETERMINE THE IMPORTANCE TO THE MODEL OF EACH FEATURE IN THE DATASET 
rfreg.fit(X,y)
pd.DataFrame({'feature':feature_cols, 'importance':rfreg.feature_importances_}).sort_values('importance', ascending=False)
````

The random forest regressor produced a result that is only slightly better than guessing:

1. nullPredictionRMSE:  2444061.99879
2. random forest RMSE_scores:  2435753.86123

Interestingly the list of feature importance did back up the intuition that aged population would be an important determinant in the healthcare spend in a zip code.

### Possible extensions of this project
Given that the modelling was never really that much better than guessing next steps should focus on improving the modeling. One immediate next step would be acquire additional data - [population density and unemployment rates](https://blog.splitwise.com/2014/01/06/free-us-population-density-and-unemployment-rate-by-zip-code/) are available by zip code and may be more predicitive of heathcare spending than other demographic features.

Another approach to improving the modeling would be to change the algorithm. Support vector regression modeling with tuning via kernel functions may yield a better fit than the approches tried here. Its features include working well in high dimensional spaces so it may interpret all the features more usefully.

Yet another approach would be engineer more features. These could include coarser age ranges, general area in the US such as Mid West, Mountains, South, Northeast etc. Other features might include proportion of aged population or if the zip code had higher than the median count of 85 year olds and over.

Alternatively the problem could be changed slightly, rather than attempting to estimate a continous number it might still be useful to try classifying the top 1/3 of zip codes as high spending and bottom 2/3 as low spending then use a logistic regressor or other classifier to just predict the group rather than an exact value. This binary classification approach would have performance monitored via AOC AUC.
